```python
inseq.list_feature_attribution_methods()
```

```python
from core.model import ModelManager
from returns.result import Success, Failure

model_config = {
    "llm_id": "meta-llama/Llama-3.2-3B",
    "device": "cpu"
}

model_result = ModelManager.initialize(model_config)
```

```python

match model_result:
    case Success(manager):
        model = inseq.load_model(manager.model, "discretized_integrated_gradients", tokenizer=manager.tokenizer)
        model.attribute(
            "Hello ladies and"
        ).show()
    case Failure(error):
        raise RuntimeError(f"Failed to load model: {error}")
```
